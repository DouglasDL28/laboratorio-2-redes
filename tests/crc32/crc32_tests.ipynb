```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("./crc32.csv")
columns = ["noise_prob","msg_len","algorithm","error","overhead"]
```

```python
ax = sns.barplot(x='noise_prob', y='error', data=dataset[dataset.msg_len > 0])
ax.set(title="CRC-32", xlabel='Noise Probability', ylabel='Error')
```

```python
print("Porcentajes de error para una probabilidad de ruido de 0.0001: ", 
      len(dataset[(dataset.msg_len > 0) & (dataset.noise_prob == 0.0001) & (dataset.error == True)]) 
      / len(dataset[(dataset.msg_len > 0) & (dataset.noise_prob == 0.0001)]))
print("Porcentajes de error para una probabilidad de ruido de 0.0005: ", 
      len(dataset[(dataset.msg_len > 0) & (dataset.noise_prob == 0.0005) & (dataset.error == True)]) 
      / len(dataset[(dataset.msg_len > 0) & (dataset.noise_prob == 0.0005)]))
print("Porcentajes de error para una probabilidad de ruido de 0.001: ", 
      len(dataset[(dataset.msg_len > 0) & (dataset.noise_prob == 0.001) & (dataset.error == True)]) 
      / len(dataset[(dataset.msg_len > 0) & (dataset.noise_prob == 0.001)]))
print("Porcentajes de error para una probabilidad de ruido de 0.005: ", 
      len(dataset[(dataset.msg_len > 0) & (dataset.noise_prob == 0.005) & (dataset.error == True)]) 
      / len(dataset[(dataset.msg_len > 0) & (dataset.noise_prob == 0.005)]))
print("Porcentajes de error para una probabilidad de ruido de 0.01: ", 
      len(dataset[(dataset.msg_len > 0) & (dataset.noise_prob == 0.01) & (dataset.error == True)]) 
      / len(dataset[(dataset.msg_len > 0) & (dataset.noise_prob == 0.01)]))
```

```python
ax = sns.histplot(x='msg_len', hue='error', data=dataset[(dataset.msg_len > 0)]);
ax.set(title='CRC-32', xlabel='Message Lenght', ylabel='Error');
```

```python
ax = sns.catplot(x='error', y='overhead', data=dataset[(dataset.msg_len > 0) & (dataset.overhead > 0)], kind='box', whis=10, aspect = 2)
ax.set(title='CRC-32', xlabel='Error', ylabel='Overhead');
```
