```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("./hamming.csv")
columns = ["noise_prob","msg_len","algorithm","error","error_corrected","error_detected","redundancy",]
```

```python
ax = sns.barplot(x='noise_prob', y='error', data=dataset)
ax.set(title="Hamming", xlabel='Noise Probability', ylabel='Error')
```

```python
ax = sns.histplot(x='msg_len', hue='error', data=dataset);
ax.set(title='Hamming', xlabel='Message Lenght');
```

```python
ax = sns.histplot(x='msg_len', hue='error_corrected', data=dataset);
ax.set(title='Hamming', xlabel='Message Lenght');
```

```python
ax = sns.histplot(x='msg_len', hue='error_detected', data=dataset);
ax.set(title='Hamming', xlabel='Message Lenght');
```
